```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import spacy
from spacy.lang.en.stop_words import STOP_WORDS as spacy_stopwords
import joblib
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nlp = spacy.load('en_core_web_sm', disable=['ner'])
```

```python
stop_words = set(stopwords.words('english')).union(spacy_stopwords)
def preprocess_text(text, lemmatize=False):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    if lemmatize:
        doc = nlp(' '.join(tokens))
        tokens = [token.lemma_ for token in doc]
    return tokens
def process_data(df, lemmatize=False):
    df['processed_review'] = df['review'].apply(lambda x: preprocess_text(x, lemmatize=lemmatize))
    df['processed_review_str'] = df['processed_review'].apply(lambda x: ' '.join(x))
    return df
```

```python
df_tokenized = process_data(df.copy(), lemmatize=False)
joblib.dump(df_tokenized, 'data_tokenized.joblib')
df_lemmatized = process_data(df.copy(), lemmatize=True)
joblib.dump(df_lemmatized, 'data_lemmatized.joblib')
```

```python
high_rating = df[df['rating'] >= 4]
low_rating = df[df['rating'] <= 2]
```

```python
high_rating['processed_review'] = high_rating['review'].apply(lambda x: preprocess_text(x))
low_rating['processed_review'] = low_rating['review'].apply(lambda x: preprocess_text(x))
```

```python
from nltk import FreqDist
high_rating_words = [word for review in high_rating['processed_review'] for word in review]
high_rating_freq = FreqDist(high_rating_words)
low_rating_words = [word for review in low_rating['processed_review'] for word in review]
low_rating_freq = FreqDist(low_rating_words)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_top_words(freq_dist, title):
    top_words = freq_dist.most_common(20)
    words, counts = zip(*top_words)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(counts), y=list(words))
    plt.title(title)
    plt.xlabel('Frequency')
    plt.ylabel('Words')
    plt.show()
plot_top_words(high_rating_freq, 'Top 20 Words in High-Rated Reviews')
plot_top_words(low_rating_freq, 'Top 20 Words in Low-Rated Reviews')
```

```python
common_words = set([word for word, freq in high_rating_freq.most_common(50)]) & set([word for word, freq in low_rating_freq.most_common(50)])
print("Common words:", common_words)
custom_stop_words = stop_words.union(common_words)
```

```python
def preprocess_text_custom(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in custom_stop_words]
    doc = nlp(' '.join(tokens))
    return [token.lemma_ for token in doc]
df['processed_review_custom'] = df['review'].apply(preprocess_text_custom)
```

```python
from wordcloud import WordCloud
def create_wordcloud(freq_dist, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()
create_wordcloud(high_rating_freq, 'Word Cloud for High-Rated Reviews')
create_wordcloud(low_rating_freq, 'Word Cloud for Low-Rated Reviews')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
df['sentiment_scores'] = df['review'].apply(lambda x: sia.polarity_scores(x))
df['neg'] = df['sentiment_scores'].apply(lambda x: x['neg'])
df['neu'] = df['sentiment_scores'].apply(lambda x: x['neu'])
df['pos'] = df['sentiment_scores'].apply(lambda x: x['pos'])
df['compound'] = df['sentiment_scores'].apply(lambda x: x['compound'])
high_rating = df[df['rating'] >= 4]
low_rating = df[df['rating'] <= 2]
high_rating_sentiment = high_rating[['neg', 'neu', 'pos', 'compound']].mean()
low_rating_sentiment = low_rating[['neg', 'neu', 'pos', 'compound']].mean()
print("High-Rated Reviews Sentiment Scores:\n", high_rating_sentiment)
print("\nLow-Rated Reviews Sentiment Scores:\n", low_rating_sentiment)
```

```python
import matplotlib.pyplot as plt
sentiment_df = pd.DataFrame({
    'High-Rated': high_rating_sentiment,
    'Low-Rated': low_rating_sentiment})
sentiment_df.plot(kind='bar', figsize=(10, 6))
plt.title('Sentiment Scores for High-Rated and Low-Rated Reviews')
plt.ylabel('Average Sentiment Score')
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
nltk.download('punkt')
nltk.download('stopwords')
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    return tokens
df['tokens'] = df['review'].apply(preprocess_text)
high_rating_tokens = df[df['rating'] >= 4]['tokens'].tolist()
low_rating_tokens = df[df['rating'] <= 2]['tokens'].tolist()
```

```python
def get_bigrams(tokens_list):
    bigram_finder = BigramCollocationFinder.from_documents(tokens_list)
    bigram_measures = BigramAssocMeasures()
    scored_bigrams = bigram_finder.score_ngrams(bigram_measures.pmi)
    return sorted(scored_bigrams, key=lambda x: x[1], reverse=True)
high_rating_bigrams = get_bigrams(high_rating_tokens)
low_rating_bigrams = get_bigrams(low_rating_tokens)
print("Top Bigrams for High-Rated Reviews:")
print(high_rating_bigrams[:10])
print("\nTop Bigrams for Low-Rated Reviews:")
print(low_rating_bigrams[:10])
```
